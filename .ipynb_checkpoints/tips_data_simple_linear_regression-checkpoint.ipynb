# Linear Regression


#### Import libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#### Import data


```python
# Get dataset
df_tips = sns.load_dataset('tips')
df_tips.head()
```

#### Distribution

```python
# Data distribution
plt.title('Tips Distribution Plot')
sns.distplot(df_tips['tip'])
plt.show()
```

```python
# Relationship between Salary and Experience
plt.scatter(df_tips['total_bill'], df_tips['tip'], color = 'lightcoral')
plt.title('Total Bills vs Tips')
plt.xlabel('Total Bills')
plt.ylabel('Tips')
plt.box(False)
plt.grid()
plt.show()
```

### Split data

##### Split into Independent/Dependent variables


```python
# Splitting variables
X = df_tips.iloc[:, :1]  # independent
y = df_tips.iloc[:, 1:2]  # dependent
```

```python
X
```

```python
y
```

```python
X.shape
```

#### Split into Train/Test sets


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Predict Results


```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
df_tips.sample(5)
```

```python
y_pred_test
```

```python
y_pred_train
```

```python
print(regressor.intercept_) # Intercept
```

```python
print(regressor.coef_) # Coefficient
```

```python
print('y = ', regressor.intercept_, '+', regressor.coef_, '* X')
print('y = ', regressor.coef_, '* X', '+', regressor.intercept_)
```

```python
print('MSE = ', mean_squared_error(y_test, y_pred_test ))
print('R2 = ', r2_score(y_test, y_pred_test ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_test )))
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.title('Total Bills vs Tips (Test Set)')
plt.xlabel('Total Bills')
plt.ylabel('Tips')
#plt.plot(X_train, y_pred_train, marker='o', color = 'firebrick')
plt.plot(X_train, y_pred_train, color = 'blue')
plt.legend(['X_train/y_train', 'X_train/Pred(y_train)'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'blue')
plt.title('Total Bills vs Tips (Test Set)')
plt.xlabel('Total Bills')
plt.ylabel('Tips')
#plt.plot(X_test, y_pred_test, marker='o', color = 'firebrick')
plt.plot(X_test, y_pred_test, color = 'lightcoral')
plt.legend(['X_test/y_test', 'X_test/Pred(y_test)'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
